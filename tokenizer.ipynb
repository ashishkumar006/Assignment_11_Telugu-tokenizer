```python
def get_stats(ids):
    counts = {}
    for pair in zip(ids, ids[1:]):
        counts[pair] = counts.get(pair, 0) + 1
    return counts

def merge(ids, pair, idx):
  newids = []
  i = 0
  while i < len(ids):
    if i < len(ids) - 1 and ids[i] == pair[0] and ids[i+1] == pair[1]:
      newids.append(idx)
      i += 2
    else:
      newids.append(ids[i])
      i += 1
  return newids
```

```python
# ---
vocab_size = 4500# the desired final vocabulary size
num_merges = vocab_size - 256
ids = list(tokens) # copy so we don't destroy the original list

merges = {} # (int, int) -> int
for i in range(num_merges):
  stats = get_stats(ids)
  pair = max(stats, key=stats.get)
  idx = 256 + i
  print(f"merging {pair} into a new token {idx}")
  ids = merge(ids, pair, idx)
  merges[pair] = idx
```

```python
vocab = {idx: bytes([idx]) for idx in range(256)}
for (p0, p1), idx in merges.items():
    vocab[idx] = vocab[p0] + vocab[p1]

def decode(ids):
  # given ids (list of integers), return Python string
  tokens = b"".join(vocab[idx] for idx in ids)
  text = tokens.decode("utf-8", errors="replace")
  return text

print(decode([128]))
```

```python
def encode(text):
  # given a string, return list of integers (the tokens)
  tokens = list(text.encode("utf-8"))
  while len(tokens) >= 2:
    stats = get_stats(tokens)
    pair = min(stats, key=lambda p: merges.get(p, float("inf")))
    if pair not in merges:
      break # nothing else can be merged
    idx = merges[pair]
    tokens = merge(tokens, pair, idx)
  return tokens
print((encode("తెలుగు")))


```

```python
import json
def decode_if_bytes(data):
    """Returns decoded string if data is bytes, else returns the original data."""
    if isinstance(data, bytes):
        try:
            # Decode from UTF-8
            return data.decode('utf-8', errors='ignore')  # Ignore decoding errors
        except UnicodeDecodeError:
            # In case decoding fails, return the raw data (binary)
            return data
    return data  # Return as is if it's not bytes

# Decode the vocabulary keys and values (if they are bytes)
vocab = {
    decode_if_bytes(key): decode_if_bytes(value) for key, value in vocab.items()
}

# Save the token-to-character mapping
with open(r"C:\Users\AISHWARYA\Downloads\New folder (7)\tokenizer_2.json", "w", encoding="utf-8") as f:
    json.dump(vocab, f, ensure_ascii=False, indent=4)



```
